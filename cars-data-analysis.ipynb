```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing DataSet

```python
cars_data = pd.read_csv("/kaggle/input/cars-dataset/2. Cars Data1.csv")
```

```python
cars_data.head()
```

```python
cars_data.shape
```

# Checking for Null Values

```python
cars_data.isnull().sum()
```

# Cleaning The Data

```python
cars_data.dropna(how = "all",inplace = True)
```

```python
cars_data.isnull().sum()
```

```python
cars_data['Cylinders'].fillna(cars_data['Cylinders'].mean(), inplace = True)
```

```python
cars_data.isnull().sum()
```

# #Checking the Manufactures & Count of Each one

```python
cars_data.head()
```

```python
cars_data['Make'].value_counts()
```

```python
car_make_counts = cars_data['Make'].value_counts()
plt.bar(car_make_counts.index, car_make_counts.values)
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.title('Number of Cars by Make')
plt.show()
```

```python
car_make_counts = cars_data['Make'].value_counts()
plt.pie(car_make_counts.values, labels=car_make_counts.index)
plt.title('Number of Cars by Make')
plt.show()
```

# Checking The Origins

```python
cars_data['Origin'].value_counts()
```

```python
cars_origin = cars_data['Origin'].value_counts()
```

```python
plt.bar(cars_origin.index, cars_origin.values)
plt.xlabel('Car Origin')
plt.ylabel('Count')
plt.title('Number of Cars by Make Origin')
plt.show()
```

# Checking The Horsepower

```python
cars_data['Horsepower'].value_counts()
```

```python
cars_hoursepower = cars_data['Horsepower'].value_counts()
plt.bar(cars_hoursepower.index, cars_hoursepower.values)
plt.xlabel('Car HoursePower')
plt.ylabel('Count')
plt.title('Number of Cars by Horsepower')
plt.show()
```

```python
plt.hist(cars_data['Horsepower'], bins=10)
plt.xlabel('Horsepower')
plt.ylabel('Count')
plt.title('Distribution of Car Horsepower')
plt.show()
```

# Checking The Cars With The Horsepower Ranging From 400 To 500

```python
hp_range = cars_data[(cars_data['Horsepower'] >= 400) & (cars_data['Horsepower'] <= 500)]
car_names = hp_range['Make'] + ' ' + hp_range['Model']
print(car_names)
```

```python
car_counts = hp_range['Make'].value_counts()

plt.bar(car_counts.index, car_counts.values)
plt.xlabel('Car Make')
plt.ylabel('Number of Cars')
plt.title('Number of Cars with Horsepower between 400 and 500')
plt.show()
```
