- We have `8830` values in `M` units

- We have `1695` value in `Varies with device` 

> Let's convert the `M` and `K` units into bytes and then remove the `M` and `K` from the values and convert them into numeric data type.

```python
# convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
# this function will convert the size column to numeric
def convert_size(size):
    # add function details here
    '''
    This function will convert the size column to numeric by multiplying the values with 1024 if it has 'k' in it and 1024*1024 if it has 'M' in it
    '''
    
    if isinstance(size, str):
        if 'k' in size:
            return float(size.replace('k', '')) * 1024
        elif 'M' in size:
            return float(size.replace('M', '')) * 1024 * 1024
        elif 'Varies with device' in size:
            return np.nan
    return size

df['Size'] = df['Size'].apply(convert_size)
```

```python
# rename the column name 'Size' to 'Size_in_bytes'
df.rename(columns={'Size': 'Size_in_bytes'}, inplace=True)
```

```python
# making a new column called 'Size in Mb' which will have the size in MB
df['Size_in_Mb'] = df['Size_in_bytes'].apply(lambda x: x/(1024*1024))
```

- Now we have converted every value into bytes and removed the `M` and `K` from the values and converted them into numeric data type.
- 'Varies with device' was a string value, therefore we intentionally converted them into null values, which we can fill later on according to our needs.

    <h1 style="color:#FF7F50; font-weight: bold;">Q-13 The relationship between the number of app listings and years, identifying the year with the most app listings?</h1>

```python
# Extract the year from the 'Last Updated' column
df['Year'] = pd.DatetimeIndex(df['Last Updated']).year

# Count the number of app listings for each year
app_count_by_year = df['Year'].value_counts().sort_index()

# Determine the year with the most app listings
year_with_most_listings = app_count_by_year.idxmax()

# Create a bar plot
plt.bar(app_count_by_year.index, app_count_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of App Listings')
plt.title('Number of App Listings by Year')
plt.show()

print("Year with the most app listings:", year_with_most_listings)
```

<h2 style="color: #FF7F50;">Observation</h2>

---
Year with the most app listings: 2018

<h2 style="color: #FF7F50;">Category</h2>

```python
# Group the dataset by year and category and count the number of app listings
app_count_by_category_year = df.groupby(['Year', 'Category']).size().unstack()

# Select a subset of categories to plot
categories_to_plot = app_count_by_category_year.columns[:5]  # Adjust the number as per your preference

# Create a line plot for the selected categories
for category in categories_to_plot:
    plt.plot(app_count_by_category_year.index, app_count_by_category_year[category], label=category)

plt.xlabel('Year')
plt.ylabel('Number of App Listings')
plt.title('Number of App Listings by Year (Category)')
plt.legend()
plt.show()
```

<h2 style="color: #FF7F50;">Observation</h2>

---

* Top `Category` with most number of apps listing are:
  * Art and Design
  * Auto and vehicles
  * Beauty
  * Books and reference
  * Business

* `Business` is booming and growing up since 2016 to 2018, on second `Books and reference` which is from 2015 to 2018

<h2 style="color: #FF7F50;">Genre</h2>

```python
# Group the dataset by year and genre and count the number of app listings
app_count_by_genre_year = df.groupby(['Year', 'Genres']).size().unstack()

# Select a subset of genres to plot
genres_to_plot = app_count_by_genre_year.columns[:5]  # Adjust the number as per your preference

# Create a line plot for the selected genres
for genre in genres_to_plot:
    plt.plot(app_count_by_genre_year.index, app_count_by_genre_year[genre], label=genre)

plt.xlabel('Year')
plt.ylabel('Number of App Listings')
plt.title('Number of App Listings by Year (Genre)')
plt.legend()
plt.show()
```

  * Adventure, Brain Games
