# Subjects With Their Number of Counts

```python
my_data['subject'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
subject_counts = my_data['subject'].value_counts()

# create the bar chart
plt.bar(subject_counts.index, subject_counts.values)

# format x-axis labels
plt.xticks(rotation=45, ha='right')

# set chart title and axis labels
plt.title("Subjects/Counts")
plt.xlabel('Subjects')
plt.ylabel('Counts')

# show the chart
plt.show()
```

# Levels & Their Counts

```python
my_data['level'].value_counts()
```

```python
level_counts = my_data['level'].value_counts()

# create the bar chart
plt.bar(level_counts.index, level_counts.values)

# format x-axis labels
plt.xticks(rotation=45, ha='right')

# set chart title and axis labels
plt.title("Level/Counts")
plt.xlabel('Level')
plt.ylabel('Counts')

# show the chart
plt.show()
```

# Checking Top 10 Course based on Number Of Subscribers

```python
# sort data by number of subscribers in ascending order and select top 10 courses
top_10 = my_data.sort_values(by='num_subscribers', ascending=False).head(10)

# display course name and number of subscribers for top 10 courses
top_10[['course_title', 'subject', 'num_subscribers']]
```

# Count Number of Courses That Are Paid and Free

```python
# count number of courses that are paid and free
paid_count = my_data[my_data['is_paid'] == True]['course_id'].count()
free_count = my_data[my_data['is_paid'] == False]['course_id'].count()

print(paid_count, free_count)
# create bar chart
plt.bar(['Paid', 'Free'], [paid_count, free_count])

# set chart title and axis labels
plt.title("Number of Paid and Free Courses")
plt.xlabel('Course Type')
plt.ylabel('Number of Courses')

# display plot
plt.show()
```

# Analysis of Udemy Course Dataset using Python Libraries Pandas and Matplotlib

In this analysis, the Udemy Course Dataset was examined using the Python libraries Pandas and Matplotlib.

Firstly, the dataset was imported and the first 20 rows were checked. Then, null values were examined.

Next, the dataset was explored to identify the different subjects and the number of courses available in each subject. It was found that there are four courses available in the dataset: Web Development (1200), Business Finance (1199), Musical Instruments (680), and Graphic Design (603). It was also observed that Web Development is the most popular subject, with the most number of courses available.

The level of the courses available in the dataset was also analyzed. It was found that the majority of the courses (1271) are at the beginner level, followed by intermediate and expert levels.

The top 10 courses were identified based on the number of subscribers using the 'num_subscribers' column. It was found that all of the top 10 courses were from Web Development, except for one course from Musical Instruments. The most popular course was 'Learn HTML5 Programming From Scratch' from Web Development with 268,923 subscribers.

The dataset was also examined to determine the number of free and paid courses available. It was found that there are 310 free courses and 3,372 paid courses.

Overall, this analysis provides insights into the Udemy Course Dataset and may be helpful for individuals looking to launch or create a course in a particular subject.
