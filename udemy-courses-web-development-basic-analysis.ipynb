# Importing **Pandas** and **Matplotlib**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Read_CSV

```python
my_data = pd.read_csv("/kaggle/input/udemy-courses-revenue-generation-and-course-anal/3.1-data-sheet-udemy-courses-web-development.csv")
```

```python
my_data.head(n=10)
```

# Show all the courses which are Free of Cost

```python
my_data[my_data.price == 0.0]
```

# Which are Top Selling Courses ?

```python
my_data.sort_values('num_subscribers', ascending=False)
```

# Which are Least Selling Courses ?

```python
my_data.sort_values('num_subscribers')
```

# Courses published years

```python
my_data['published_timestamp'] = pd.to_datetime(my_data.published_timestamp)
```

```python
my_data['Year'] = my_data['published_timestamp'].dt.year
```

```python
year_count = my_data.groupby('Year').size()
```

```python
year_count.plot(kind='bar')
plt.show()
```

# What is the Max. Number of Subscribers for Each Level of courses ?

```python
my_data.groupby('level')['num_subscribers'].max()
```

```python
max_no_sub = my_data.groupby('level')['num_subscribers'].max()
```

```python
max_no_sub.plot(kind='bar')
plt.show()
```

# What is the Min. Number of Subscribers for Each Level of courses ?

```python
my_data.groupby('level')['num_subscribers'].min()
```

```python
min_no_sub = my_data.groupby('level')['num_subscribers'].min()
```

```python
min_no_sub.plot(kind='bar')
plt.show()
```
