```python
data = pd.read_csv("/kaggle/input/superhero-dataset/cleaned_superhero_dataset.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.replace(np.nan, None, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(10)
```

```python
data['origin'].value_counts()
```

```python
import matplotlib.pyplot as plt

origin = data['origin'].value_counts()
```

```python
# create bar chart
plt.bar(origin.index, origin.values)

# add labels and title
plt.xlabel('Origin')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Value')
plt.title('Bar Chart Example')

# display plot
plt.show()
```

```python
publisher = data['publisher'].value_counts().head(10)
```

```python
# create bar chart
plt.bar(publisher.index, publisher.values)

# add labels and title
plt.xlabel('Publisher')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Value')
plt.title('Bar Chart Example')

# display plot
plt.show()
```
