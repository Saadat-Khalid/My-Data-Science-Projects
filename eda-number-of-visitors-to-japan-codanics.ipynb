## Which Country visits to Japan more each Year?

```python
df.columns
```

```python
top_countries_by_year = df.groupby(['Year', 'Country/Area'])['Visitor']
top_countries_by_year.sum().groupby('Year').nlargest(5)
```

```python
top_countries_by_year = df.groupby(['Year', 'Country/Area'])['Visitor'].sum().reset_index()
top_countries_by_year = top_countries_by_year.groupby('Year').apply(lambda x: x.nlargest(10, 'Visitor')).reset_index(drop=True)

fig = px.bar(top_countries_by_year, x='Year', y='Visitor', color='Country/Area', title='Top 10 Countries by Visitor Count')
fig.show()
```

## Observation:
___

- China has consistently been the largest source of visitors to Japan from 2017 to 2023, except for a significant decrease in 2021.
- South Korea and Taiwan are the second and third most common countries of origin for visitors to Japan, following China.
- Hong Kong and the United States are also among the top five countries of origin for visitors to Japan.
- The number of visitors from South Korea to Japan has shown a significant increase over the years.
- Taiwan has had a stable number of visitors to Japan, with a slight increase in 2023.
- The United States has experienced a gradual increase in visitor count to Japan, reaching its peak in 2023.
- Hong Kong has shown fluctuating visitor numbers, with a notable decrease in 2023.
- Other countries such as Thailand, Vietnam, and Nepal also contribute to the visitor count to Japan, but in smaller numbers compared to the top countries mentioned above.


## Which Country Least Visits to Japan each Year?

```python
lowest_countries_by_year = df.groupby(['Year', 'Country/Area'])['Visitor']
lowest_countries_by_year.sum().groupby('Year').nsmallest(10)
```

```python
lowest_countries_by_year = df.groupby(['Year', 'Country/Area'])['Visitor'].sum().reset_index()
lowest_countries_by_year = lowest_countries_by_year.groupby('Year').apply(lambda x: x.nsmallest(10, 'Visitor')).reset_index(drop=True)

fig = px.bar(lowest_countries_by_year, x='Year', y='Visitor', color='Country/Area', title='Least 10 Countries by Visitor Count')
fig.show()
```

## Observation:
____

- East Germany, Serbia and Montenegro, South Yemen, and Unknown had no recorded visitors to Japan from 2017 to 2023, indicating minimal or no tourist exchange with Japan during this period.
- Central Africa, Equatorial Guinea, Sao Tome and Principe, and Vatican had consistently low visitor numbers to Japan, suggesting a relatively small number of tourists from these countries.
- Guinea-Bissau showed a small but increasing interest in visiting Japan, with visitor counts ranging from 13 in 2017 to 30 in 2019 and 6 in 2023.
- North Korea had a small number of visitors to Japan, with counts ranging from 13 in 2018 and 2019 to 0 in subsequent years, suggesting potential travel restrictions or limitations.
- Saint Vincent, Micronesia, Belize, Nauru, San Marino, and Saint Kitts and Nevis recorded a small number of visitors to Japan in specific years, indicating limited tourist exchange with Japan.


## How many Visitors Arrival Over Year?

```python
df_year.groupby('Year')['Visitor Arrivals'].sum()
```

```python
fig = px.scatter(df_year, x='Year', y='Visitor Arrivals', size = 'Visitor Arrivals', color='Visitor Arrivals' ,title='Visitor Arrivals to Japan Over Time')
fig.update_layout(xaxis_title='Year', yaxis_title='Visitor Arrivals')

# Show the plot
fig.show()
```

## Observation:
___


- The number of visitor arrivals to Japan has generally increased over the years.
- There was a significant increase starting from the 1970s, with a peak of 24 million arrivals in 2016.
- 2017 saw a further increase to nearly 29 million arrivals.
- In 2018 and 2019, there were over 31 million arrivals each year.
- However, due to the COVID-19 pandemic, the number dropped to around 4 million in 2020.
- In 2021, there were only approximately 246,000 arrivals.
- The years 2022 and 2023 show signs of recovery with 3.8 million and 10.7 million arrivals, respectively.
